```python
# define predictor variables and the response variable
y = diab_df['Outcome']
X = diab_df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]

logreg_model = LogisticRegression()
logreg_model.fit(X, y)
```

```python
## 2 points

## Explain what the code is doing:
## What are the prior probabilities of the intercept and coefficients?
# Which likelihood distribution has been used?
## What does pm.invlogit(linreg) mean?
## What is map_est?

with pm.Model() as logreg_model:
    # random variables for coefficients with
    # uninformative priors for each parameter

    w0 = pm.Normal('w0', mu=0, sd=100)
    w1 = pm.Normal('w1', mu=0, sd=100)
    w2 = pm.Normal('w2', mu=0, sd=100)
    w3 = pm.Normal('w3', mu=0, sd=100)
    w4 = pm.Normal('w4', mu=0, sd=100)
    w5 = pm.Normal('w5', mu=0, sd=100)
    w6 = pm.Normal('w6', mu=0, sd=100)
    w7 = pm.Normal('w7', mu=0, sd=100)
    w8 = pm.Normal('w8', mu=0, sd=100)
   
    linreg = w0 * np.ones(diab_df.shape[0]) + w1 * diab_df.Pregnancies.values + w2 * diab_df.Glucose.values \
    + w3 * diab_df.BloodPressure.values + w4 * diab_df.SkinThickness.values + w5 * diab_df.Insulin.values + \
    w6 * diab_df.BMI.values + w7 * diab_df.DiabetesPedigreeFunction.values + w8 * diab_df.Age.values
    
    p_outcome = pm.invlogit(linreg)

    # Transform random variables into vector of probabilities p(y_i=1)
    # according to logistic regression model specification.
    likelihood = pm.Bernoulli('likelihood', p_outcome, observed=diab_df.Outcome.values)

    # Bernoulli random vector with probability of success
    # given by sigmoid function and actual data as observed
    map_est= pm.find_MAP()
    print(map_est)
```

### Explain what the code is doing:
We are building a Bayesian logistic regression model, so we have to put a prior distribution on each variable. Since we have no prior knowledge on the parameters, we assume a Normal distribution(Gaussian distribution) with mean as 0 and Standard Deviation of 100. 

image.png

Once we get the set of all variables [beta0 and beta(var)] into a variable (here 'linreg') we use 'logit' function.

The likelihood combines the parameters with the data according to the specification of logistic regression.

Our target variable (Outcome) is modelled as a Bernoulli random variable with success probability given by likelihood.

### What are the prior probabilities of the intercept and coefficients?
 The prior probabilities are estimated and are calculated based on probabilities function that follows a known distribution. These are Uninformative priors for each parameter. 

### Which likelihood distribution has been used?
Bernoulli distribution. For Bernoulli distribution there is only single parameter p. 

### What does pm.invlogit(linreg) mean?
Inverse-logit function, transforms continuous values to the range (0, 1).
logit(x)
invlogit(x)
where x is vector of continuous values.

### What is map_est?
MAP (Maximum A Posteriori) estimate. The MAP estimate of 'X' given Y=y is the value of x that maximizes the posterior PDF or PMF. In logistic regression we find best prameters by MAP solution.
